```python
import requests
from requests import Response

url:str = "https://raw.githubusercontent.com/Chi-Feng-Tseng/test/refs/heads/main/lesson10/students.csv"

# 發送GET 請求並取得網頁內容
response:Response = requests.get(url)

# 印出狀態碼，200 表示成功
if response.status_code == 200:
    print("下載成功")
    content:str = response.text
```

```python
print(content)
```

```python
import io
from io import StringIO
import csv

file = io.StringIO(content)
reader = csv.DictReader(file)
students = [row for row in reader]
file.close()
students
```
