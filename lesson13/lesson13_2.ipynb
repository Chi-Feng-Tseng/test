```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, label='line plot')
axes.set_title('Line plot on Existing Axes')  #將圖表的title放上去
axes.legend()  #圖表的左上角多了'line plot'的顯示
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager  #要顯示中文，需多打這兩行
import matplotlib as mpl

fontManager.addfont('chineseFont.ttf')
mpl.rc('font', family = 'chineseFont')

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, marker='o',label='學生1',linestyle='-.',color='r',markersize=10,linewidth=2)
axes.set_title('折線圖')  
axes.legend()  
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager  #要顯示中文，需多打這兩行
import matplotlib as mpl

fontManager.addfont('chineseFont.ttf')
mpl.rc('font', family = 'chineseFont')

figure, axes = plt.subplots()
x = [1, 2, 3]
y = [3, 4, 5]
axes.plot(x, y, marker='o',label='學生1',linestyle='-.',color='r',markersize=10,linewidth=2)
axes.set_title('折線圖') 
axes.set_xticks([1, 2, 3],['第1','第2','第3'])  #這裡的1, 2, 3要對應X的1, 2, 3
axes.set_yticks([3, 4, 5],['3', '4','5'])   #這裡的3, 4, 5要對應y的31, 4, 5
axes.legend()  
plt.show()
```
