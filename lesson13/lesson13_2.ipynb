```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager  #
import matplotlib as mpl

fontManager.addfont('chineseFont.ttf')
mpl.rc('font', family = 'chineseFont')

figure, axes = plt.subplots()
x = [1, 2, 3]
y = [3, 4, 5]
axes.plot(x, y, marker='o',label='1',linestyle='-.',color='r',markersize=10,linewidth=2)
axes.set_title('') 
axes.set_xticks([1, 2, 3],['1','2','3'])  #1, 2, 3X1, 2, 3
axes.set_yticks([3, 4, 5],['3', '4','5'])   #3, 4, 5y31, 4, 5
axes.legend()  
plt.show()
```
