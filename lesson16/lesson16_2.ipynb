```python
scores = np.random.randint(50, 101 ,size=(50, 5))
score_pd = pd.DataFrame(scores, 
                         columns=['國文','英文','數學','地理','歷史'],
                         index=[f'學生{index}' for index in range(1, 51)])
score_pd.columns.name = '科目'
score_pd.index.name = '學號'
sum_series = np.sum(score_pd, axis=1)
mean_series = score_pd.mean(axis=1)
score_pd['總分'] = sum_series   #若是沒打這兩行，是不會跑出總分與平均的欄位
score_pd['平均'] = mean_series
```

```python
score_pd
```

```python
#排序
score_pd.sort_values(by='總分', ascending=False)    #ascending 為True就是由小排到大，若為False則是由大排到小
```

```python
sorted_pd = score_pd.sort_values(by='總分', ascending=False)
sorted_pd['總分']
```

```python
sorted_pd = score_pd.sort_values(by='總分', ascending=False)
sorted_pd['總分'].rank(ascending=False,method='min') #若有同名次，則取最小值
```

```python
sorted_pd = score_pd.sort_values(by='總分', ascending=False)
rand_series = sorted_pd['總分'].rank(ascending=False,method='min')
sorted_pd['名次'] = rand_series
```
